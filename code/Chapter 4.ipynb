The methodology to solve this kind of problem:

1. Propose hypotheses based on possbile values
2. Assign likelihood function (important), defaulting uniform distribution unless extra information provided
3. Update the Probability Mass Function (PMF) based on the events happened
4. Output the hypothesis with the maximum likelihood, mean, median and confidence interval of the PMF

```python
from thinkbayes import Suite, Percentile, CredibleInterval, Beta
import thinkplot
```

```python
class Euro(Suite):
    def Likelihood(self, data, hypo):
        head, tail = data
        x = hypo / 100.0
        like = x ** head * (1 - x) ** tail
        return like
```

```python
def MaximumLikelihood(pmf):
    prob, val = max((prob, val) for val, prob in pmf.Items())
    return val
```

```python
def printstats(suite):
    print 'Max', MaximumLikelihood(suite)
    print 'Mean', suite.Mean()
    print 'Median', Percentile(suite, 50)
    print 'CI', CredibleInterval(suite, 90)
    print suite.Prob(50)
```

```python
def TrianglePrior():
    suite = Euro()
    for x in range(0,51):
        suite.Set(x,x)
    for x in range(51,101):
        suite.Set(x, 100 - x)
    suite.Normalize()
    return suite
```

```python
suite = TrianglePrior()
head, tail = 140, 110
suite.Update((head, tail))
thinkplot.pmf(suite)
thinkplot.show()
printstats(suite)
```

```python
class Beta(object):
    def __init__(self, alpha=1, beta=1):
        self.alpha = alpha
        self.beta = beta
    def Update(self, data):
        heads, tails = data
        self.alpha += heads
        self.beta += tails
```

```python
beta = Beta()
beta.Update((140,110))
print beta.Mean()
thinkplot.pmf(beta.MakePmf())
thinkplot.show()
```
